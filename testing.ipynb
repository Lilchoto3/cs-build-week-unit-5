```python
import csv
points = []
with open('blobs.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        points.append([float(row[1]), float(row[2])])

print(points[:5])
```

```python
from kmeans.kmeansClusteringNoScipy import KMeans
kmeans = KMeans(n_clusters=4, max_iter=20)
kmeans.fit(points)
print(kmeans.predict(points))
```

```python
# plot points with clusters to show it works
import pandas as pd
import matplotlib.pyplot as plt

def plot_clusters(clusters, iterations):
    kmeans = KMeans(n_clusters=clusters, max_iter=iterations)
    kmeans.fit(points)

    df = pd.DataFrame(points, columns=['x','y'])
    df['label'] = kmeans.predict(points)

    colors = {0:'tab:blue',
            1:'tab:orange', 
            2:'tab:green', 
            3:'tab:red', 
            4:'tab:purple',
            5:'tab:brown',
            6:'tab:pink', 
            7:'tab:gray'
    }
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key,                       color=colors[key])
    plt.show()

plot_clusters(4, 20) 
```

```python
# variations over iteration
%matplotlib inline
plot_clusters(4,1)
plot_clusters(4,2)
plot_clusters(4,3)
plot_clusters(4,4)
plot_clusters(4,5)
plot_clusters(4,6)
plot_clusters(4,7)
plot_clusters(4,8)
plot_clusters(4,9)
plot_clusters(4,10)
```

```python
# variations over clusters
plot_clusters(1, 20)
plot_clusters(2, 20)
plot_clusters(3, 20)
plot_clusters(4, 20)
plot_clusters(5, 20)
plot_clusters(6, 20)
plot_clusters(7, 20)
```
