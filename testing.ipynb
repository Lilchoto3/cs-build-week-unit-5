```python
import csv
points = []
true_labels = []
with open('iris.data', newline='') as f:
    reader = csv.reader(f)
    # next(reader)
    for row in reader:
        if len(row) > 0:
            points.append([float(x) for x in row[:4]])
            if row[4] == "Iris-setosa":
                true_labels.append(0)
            elif row[4] == "Iris-versicolor":
                true_labels.append(1)
            else:
                true_labels.append(2)

print(points[:5])
print(true_labels[:5])
```

```python
from kmeans.kmeansClusteringNoScipy import KMeans
kmeans = KMeans(n_clusters=3, max_iter=20)
kmeans.fit(points)
print(kmeans.predict(points))
```

```python
# plot points with clusters to show it works
import pandas as pd
import matplotlib.pyplot as plt

def plot_clusters(clusters, iterations):
    kmeans = KMeans(n_clusters=clusters, max_iter=iterations)
    kmeans.fit(points)

    df = pd.DataFrame(points,
                      columns=['sepal-length',
                               'sepal-width',
                               'petal-length',
                               'petal-width'
                               ]
                     )
    df['label'] = kmeans.predict(points)

    colors = {0:'tab:blue',
            1:'tab:orange', 
            2:'tab:green', 
            3:'tab:red', 
            4:'tab:purple',
            5:'tab:brown',
            6:'tab:pink', 
            7:'tab:gray'
    }
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax,
                   kind='scatter',
                   x='petal-length',
                   y='petal-width',
                   label=key,
                   color=colors[key])
    plt.show()

plot_clusters(3, 20) 
```

```python
# variations over iteration
%matplotlib inline
plot_clusters(3,1)
plot_clusters(3,2)
plot_clusters(3,3)
plot_clusters(3,4)
plot_clusters(3,5)
plot_clusters(3,6)
plot_clusters(3,7)
plot_clusters(3,8)
plot_clusters(3,9)
plot_clusters(3,10)
plot_clusters(3,11)
plot_clusters(3,12)
```

```python
# variations over clusters
plot_clusters(1, 20)
plot_clusters(2, 20)
plot_clusters(3, 20)
plot_clusters(4, 20)
plot_clusters(5, 20)
plot_clusters(6, 20)
plot_clusters(7, 20)
```
